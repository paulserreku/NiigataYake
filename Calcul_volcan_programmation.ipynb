On crée une fonction qui va nous permettre de trouver les écarts les plus proches de zéro, et qui va nous renvoyer une valeur positive et une valeur négative.

```python
def ecart_proche_de_zero (liste_ecart,liste_ΔV,liste_f):
    """
    On crée une fonction qui nous donne les écarts les plus proches de zéro (une valeur négative et une valeur positive)
    Ces deux valeurs seront les bornes de notre prochain intervale pour "zoomer" et obtenir des valeurs plus précises de ΔV et f
    """
    ecarts_droit = 0.0001  #on definit arbitrairement un arrondi au millimetre pres car la valeur donnee dans notre tableau est arrondi a 10-3
    ecarts_gauche = -0.0001
    valeur_ΔV=None
    valeur_f=None
    valeur_1_ΔV = None
    valeur_1_f = None
    n=0
    
    for i in liste_ecart :
        
        if i >= 0 :
            if i <= ecarts_droit:
                ecarts_droit=i
                valeur_ΔV = liste_ΔV[n]
                valeur_f = liste_f[n]
                
        if i <= 0 :
            if i >= ecarts_gauche:
                ecarts_gauche=i
                valeur_1_ΔV = liste_ΔV[n]
                valeur_1_f = liste_f[n]

        n = n+1
        
    return ecarts_gauche, ecarts_droit, valeur_ΔV,valeur_f,valeur_1_ΔV,valeur_1_f



#Cette fonction a pour but de trouver les deux valeurs les plus proches de zéro dans une liste (positive et négative)."""
#Puisqu'il y aura deux valeurs, une positive et une négative, notre fonction aura deux parties.

    
```

### Méthode 1
Pour chaque station, on tente d'approcher le meilleur couple (ΔV,f), c'est à dire celui pour lequel l'écart entre la valeur de Uz calculée et la valeur de Uz donnée par le sujet est le plus petit, se rapproche le plus de 0.

```python
# ON EFFECTUE UNE BOUCLE POUR LES 10 STATIONS

stations_data = {
    'Station': [],
    'f_1': [],
    'f_2': [],
    'ΔV_1': [],
    'ΔV_2': []
}

valeurs_ΔV = []
valeurs_f = []

for station_index in range(10):
    result1 = []
    result2 = []
    liste_ecarts_1 = []
    
    valeur_negative_1 = -0.0001
    valeur_positive_1 = 0.0001

    for ΔV in np.linspace(0, 60e6, 2000):  # Variation de volume
        for f in np.linspace(-10e3, 0, 200):  # Profondeur
            
            Uz = ((3 * ΔV * (f + stations['Elevation'][station_index])) / 
                  (4 * np.pi * (np.power((np.power((f + stations['Elevation'][station_index]),2) + np.power((stations['r'][station_index]),2)), (3 / 2)))))
        
            difference_1 = stations['U_up'][station_index] - Uz

            if valeur_negative_1 <= difference_1 <= valeur_positive_1:
                liste_ecarts_1.append(difference_1)
                result1.append(ΔV)
                result2.append(f)

    print(f"Station {station_index + 1}:")
    print(f"Les deux meilleurs couples (ΔV,f) calculés sont {(ecart_proche_de_zero(liste_ecarts_1, result1, result2)[2],ecart_proche_de_zero(liste_ecarts_1, result1, result2)[3])} et {(ecart_proche_de_zero(liste_ecarts_1, result1, result2)[4],ecart_proche_de_zero(liste_ecarts_1, result1, result2)[5])}")

    plt.plot(result1, result2, label=f"Station {station_index + 1}")

    # Affinage des valeurs pour une deuxième itération

    result3 = []
    result4 = []
    liste_ecarts_2 = []

    valeur_negative_2 = ecart_proche_de_zero(liste_ecarts_1, result1, result2)[0]
    valeur_positive_2 = ecart_proche_de_zero(liste_ecarts_1, result1, result2)[1]

    for ΔV in np.linspace(ecart_proche_de_zero(liste_ecarts_1, result1, result2)[2], 
                          ecart_proche_de_zero(liste_ecarts_1, result1, result2)[4], 2000):
        for f in np.linspace(ecart_proche_de_zero(liste_ecarts_1, result1, result2)[3], 
                             ecart_proche_de_zero(liste_ecarts_1, result1, result2)[5], 200):
            
            Uz = ((3 * ΔV * (f + stations['Elevation'][station_index])) / 
                  (4 * np.pi * (np.power((np.power((f + stations['Elevation'][station_index]),2) + np.power((stations['r'][station_index]),2)), (3 / 2)))))
            difference_2 = stations['U_up'][station_index] - Uz

            if valeur_negative_2 <= difference_2 <= valeur_positive_2:
                liste_ecarts_2.append(difference_2)
                result3.append(ΔV)
                result4.append(f)

    print(f"Premier affinage Station {station_index + 1}:")
    print(f"Les deux meilleurs couples (ΔV,f) calculés sont {(ecart_proche_de_zero(liste_ecarts_2, result3, result4)[2],ecart_proche_de_zero(liste_ecarts_2, result3, result4)[3])} et {(ecart_proche_de_zero(liste_ecarts_2, result3, result4)[4],ecart_proche_de_zero(liste_ecarts_2, result3, result4)[5])}")

    plt.plot(result3, result4, linestyle='--', label=f"Premier affiné Station {station_index + 1}")

    # Affinage des valeurs pour une deuxième itération

    result5 = []
    result6 = []
    liste_ecarts_3 = []

    valeur_negative_3 = ecart_proche_de_zero(liste_ecarts_2, result3, result4)[0]
    valeur_positive_3 = ecart_proche_de_zero(liste_ecarts_2, result3, result4)[1]

    for ΔV in np.linspace(ecart_proche_de_zero(liste_ecarts_2, result3, result4)[2], 
                          ecart_proche_de_zero(liste_ecarts_2, result3, result4)[4], 2000):
        for f in np.linspace(ecart_proche_de_zero(liste_ecarts_2, result3, result4)[3], 
                             ecart_proche_de_zero(liste_ecarts_2, result3, result4)[5], 200):
            
            Uz = ((3 * ΔV * (f + stations['Elevation'][station_index])) / 
                  (4 * np.pi * (np.power((np.power((f + stations['Elevation'][station_index]),2) + np.power((stations['r'][station_index]),2)), (3 / 2)))))
            
            difference_3 = stations['U_up'][station_index] - Uz

            if valeur_negative_3 <= difference_3 <= valeur_positive_3:
                liste_ecarts_3.append(difference_3)
                result5.append(ΔV)
                result6.append(f)

    # Ajout des valeurs ΔV et f dans les listes
    ΔV_1 = ecart_proche_de_zero(liste_ecarts_3, result5, result6)[2]
    ΔV_2 = ecart_proche_de_zero(liste_ecarts_3, result5, result6)[4]
    f_1 = ecart_proche_de_zero(liste_ecarts_3, result5, result6)[3]
    f_2 = ecart_proche_de_zero(liste_ecarts_3, result5, result6)[5]

    valeurs_ΔV.extend([ΔV_1, ΔV_2])
    valeurs_f.extend([f_1, f_2])

    # Enregistrement dans les listes pour chaque station
    stations_data['Station'].append(f"Station {station_index + 1}")
    stations_data['f_1'].append(f_1)
    stations_data['f_2'].append(f_2)
    stations_data['ΔV_1'].append(ΔV_1)
    stations_data['ΔV_2'].append(ΔV_2)



    print(f"Deuxieme Affinage Station {station_index + 1}:")
    print(f"Les deux meilleurs couples (ΔV,f) calculés sont {(ecart_proche_de_zero(liste_ecarts_3, result5, result6)[2],ecart_proche_de_zero(liste_ecarts_3, result5, result6)[3])} et {(ecart_proche_de_zero(liste_ecarts_3, result5, result6)[4],ecart_proche_de_zero(liste_ecarts_3, result5, result6)[5])}")

    plt.plot(result5, result6, linestyle='--', label=f"Second affiné Station {station_index + 1}")
    
    plt.legend()
    plt.show()
```

### Méthode 2
On calcule directement les valeurs ΔV et f à partir des données du tableau et des déplacements par rapport à la localisation du volcan.

```python
#Calcul de r, la différence horizontale de la source
stations['r']= np.sqrt((np.square(df_position_volcan['utm_nord'][0]-stations['Eastern']))+(np.square(df_position_volcan['utm_est'][0]-stations['Northern'])))

#Calcul de Ur, le déplacement horizontal, en fonction de U_Est et U_Nord
stations['Ur']=np.sqrt((np.square(stations['U_east']))+np.square(stations['U_north']))

#Calcul de f, la profondeur sous le niveau de la mer (en m)
stations['f']=((stations['r']*stations['U_up'])/stations['Ur'])-stations['Elevation']

#Calcul de la variation de volume dans le reservoir appelée ΔV (en m^3)
stations['ΔV']=(4*np.pi*(np.power(np.square(stations['r']*stations['U_up']/stations['Ur'])+np.square(stations['r']),(3/2)))*stations['Ur'])/(3*stations['r'])

#Ici la valeur de Uz est équivalent à la valeur de U_up 
#La valeur h correspond à l'elevation 

output_path = os.path.join(output_dir, 'résultats_stations_méthode2.csv')
stations.to_csv(output_path, index=False)
```
